```python
from transformers import AutoModelForCausalLM, CodeGenTokenizerFast as Tokenizer
from PIL import Image

model_id = "vikhyatk/moondream1"
model = AutoModelForCausalLM.from_pretrained(model_id, trust_remote_code=True)
tokenizer = Tokenizer.from_pretrained(model_id)
```

```python
video_id="aBZri41tUsM"
thumbnail_link = f"https://img.youtube.com/vi/{video_id}/maxresdefault.jpg"
print(thumbnail_link)

import requests
from io import BytesIO
from PIL import Image



image = Image.open(BytesIO(requests.get(thumbnail_link).content))
print(image.size)
print("DONE")
```

```python
enc_image = model.encode_image(image)
caption=model.answer_question(enc_image, "Write in details what in the picture", tokenizer)
print(caption)

data = {

    "url":thumbnail_link,
    "caption":caption
}

collection.insert_one(data)
```
